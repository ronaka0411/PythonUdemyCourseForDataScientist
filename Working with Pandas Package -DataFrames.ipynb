```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']= 8,6
%matplotlib inline
```

```python
#Distribution
import warnings
warnings.filterwarnings('ignore')

vis1 = sns.distplot(stats.InternetUsers)
```

```python
vis1 = sns.distplot(stats.InternetUsers, bins=40)
```

```python
vis2 = sns.boxplot(data=stats,x='IncomeGroup',y='BirthRate')
```

```python
# linear model plot
vis3 = sns.lmplot(data=stats, x='InternetUsers',y='BirthRate', \
                  hue='IncomeGroup', fit_reg = False, size=10)
```

```python
# Keyword Argument (Advanced Tutorial)

# change marker size
vis3 = sns.lmplot(data=stats, x='InternetUsers',y='BirthRate', \
                  hue='IncomeGroup', fit_reg = False, size=10, scatter_kws={"s":300})
```
